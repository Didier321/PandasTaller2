```python
import pandas as pd
import matplotlib.pyplot as plt



df = pd.read_excel('04_TallerLimpiezaDatos.xlsx')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.loc[df['Genero'] == 'H', 'Genero'] = 'Masculino'
df.loc[df['Genero'] == 'Hombre', 'Genero'] = 'Masculino'
df.loc[df['Genero'] == 'Mujer', 'Genero'] = 'Femenino'
df.loc[df['Genero'] == 'F', 'Genero'] = 'Femenino'
df.loc[df['Genero'] == 'M', 'Genero'] = 'Femenino'

```

```python
df.dropna(inplace= True)
df.to_string()
```

```python
df.info()
df.isnull().sum()
```

```python
df['SubTotal'] = df['Cantidad'] * df['ValorUnitario']

df
```

```python
tot_ubicacion = df.groupby('Ubicación')['Cantidad'].sum()

plt.bar(tot_ubicacion.index, tot_ubicacion.values)
plt.xlabel("Ubicación")
plt.ylabel("Cantidad")
plt.title("Cantidad Total por Ubicación")
plt.show()
```

```python
tot_genero = df.groupby('Genero')['SubTotal'].sum()
plt.bar(tot_genero.index , tot_genero.values)
plt.xlabel('Genero')
plt.ylabel('Total')
plt.title('Total Por Genero')
plt.show()
tot_genero
```

```python
df.to_csv("04_TallerLimpiezaDatos.xlsx", index=False)
```
